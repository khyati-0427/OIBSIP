SPLITTING THE DATASET INTO TRAIN AND TEST

```python
X= dataset[['Id','SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
Y= dataset['Species']
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.20,random_state=42)
# Let's create an instance for the LogisticRegression model
lr = LogisticRegression()

# Train the model on our train dataset
lr.fit(X,Y)

# Train the model with the training set

# Getting predictions from the model for the given examples
lr.fit(X_train,Y_train)
predicted = lr.predict(X)

# Compare with the actual charges

Scores = pd.DataFrame({'Actual':Y,'Predicted':predicted})
Scores.head()
```

```python
Y_test_hat=lr.predict(X_test)
print(accuracy_score(Y_test,Y_test_hat)*100,'%')
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)

knn.fit(X_train_std,Y_train)
```

```python
predict_knn=knn.predict(X_test_std)
accuracy_knn=accuracy_score(Y_test,predict_knn)*100
accuracy_knn
```
